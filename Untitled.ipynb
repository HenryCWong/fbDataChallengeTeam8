```python
import pandas as pd
import numpy as np
import seaborn as sns
import plotly.plotly as py
import plotly.graph_objs as go

sns.set()
```

```python
df = pd.read_csv("sf_business_cleaned.csv")
```

```python
df.head()
```

```python
df["lic_code"].unique()
```

```python
dfnew = df.loc[df['naics_code'] == "Private Education and Health Services"]
```

```python
dfnew
```

```python
healthCounts = dfnew["lic_code"].value_counts()
```

```python
healthCounts
```

```python
import plotly.plotly as py
import plotly.graph_objs as go

labels = healthCounts.index
values = healthCounts.values

trace = go.Pie(labels=labels, values=values)
trace["showlegend"] = False

py.iplot([trace], filename='basic_pie_chart')
```

```python
healthCountsSmall = healthCounts[:5]
```
